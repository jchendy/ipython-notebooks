```python
#Boilerplate twitter stuff; taken straight from the book

import twitter

# XXX: Go to http://dev.twitter.com/apps/new to create an app and get values
# for these credentials, which you'll need to provide in place of these
# empty string values that are defined as placeholders.
# See https://dev.twitter.com/docs/auth/oauth for more information 
# on Twitter's OAuth implementation.

CONSUMER_KEY = ''
CONSUMER_SECRET = ''
OAUTH_TOKEN = ''
OAUTH_TOKEN_SECRET = ''

auth = twitter.oauth.OAuth(OAUTH_TOKEN, OAUTH_TOKEN_SECRET,
                           CONSUMER_KEY, CONSUMER_SECRET)

twitter_api = twitter.Twitter(auth=auth)

# Nothing to see by displaying twitter_api except that it's now a
# defined variable

print twitter_api
```

```python
#initialize variables and get data from twitter

import time

queries = ['science', '"data science"', 'data', 'ipython']
statuses = dict()
status_texts = dict()

for q in queries:
    statuses[q] = []
    args = {'q': q, 'count': 100}
    
    for _ in range(50):
        search_results = twitter_api.search.tweets(**args)      
        #args['max_id'] = search_results['search_metadata']['next_results']['max_id']
        
        try:
            next = search_results['search_metadata']['next_results']
        except KeyError, e: # No more results when next_results doesn't exist
            break
            
        start = next.find('max_id=')
        next = next[start + 7:]
        end = next.find('&')
        next = next[:end]
        args['max_id'] = next
        
        statuses[q] += search_results['statuses']
        time.sleep(.2)
    print q,len(statuses[q])
        
    status_texts[q] = [ status['text'] 
                     for status in statuses[q] ]
```

```python
#generate word count from each query 

from collections import Counter
from prettytable import PrettyTable

counters = dict()

def clean(w):
    w = w.lower() 
    for c in ['.', ',', '(', ')', '!', ':', '#']:
        length = len(w)
        if length > 0 and w[0] == c:
            w = w[1:]
        length = len(w)            
        if length > 0 and w[length - 1] == c:
            w = w[:length - 1]
    return w

for q in queries:
    words = [ clean(w)
              for t in status_texts[q] 
                  for w in t.split() ]
    
    counters[q] = Counter(words)
    
    pt = PrettyTable(field_names=['Word', 'Count']) 
    [ pt.add_row(kv) for kv in counters[q].most_common()[:50] ]
    pt.align['Word'], pt.align['Count'] = 'l', 'r' # Set column alignment
```

```python
#generate list of programming languages
import re

#file is a dump of the markup from wikipedia's languages list
languages_file = open('./langs_from_wikipedia.txt', 'r')
lang_text = languages_file.read()

#things like [name of thing | link to thing]
languages = [l.lower() for l in re.findall('(?<=\|)[^\]]*(?=])', lang_text)]
#things like [[thing]]
languages += [l.lower() for l in re.findall('(?<=\[\[)[^\]|]*(?=])', lang_text)]
languages = set(languages)
```

```python
language_counts = dict()

for q in queries:
    pairs = []
    for l in languages:
        count = counters[q][l.lower()]
        if count > 0:
            pairs.append({'language': l, 'count': count})
    language_counts[q] = sorted(pairs, key = lambda x: x["count"], reverse = True)
    
for q in queries:
    pt = PrettyTable(field_names=['Language', 'Count']) 
    for row in language_counts[q]:
        pt.add_row([row['language'], row['count']])
    pt.align['Language'], pt.align['Count'] = 'l', 'r' # Set column alignment
    print q + ' -- total tweets', len(status_texts[q])
    print pt
```
